# Read Spotter

```python
import matplotlib.pyplot as plt
from wavespectra import read_spotter
```

## CSV

### 2D spectra

The spectra are read as 2D, directional spectra by default by using the mean frequency
and the mean direcional spreading data to fit a $cos^2$ distribution to the 1D, spectral
data. The bulk parameters are also returned as variables in the dataset

```python
dset = read_spotter("data/spotter_20210929.csv")
dset
```

### Direction resolution

The ``dd`` parameter can be used to define the direction resolution the 2D spectra is
constructed as. One can set it as ``None`` to return the 1D spectra as provided in the
input files:

```python
dset = read_spotter("data/spotter_20210929.csv", dd=None)
dset
```

Or use any value to determine the direction resolution of the constructed spectra:

```python
dset = read_spotter("data/spotter_20210929.csv", dd=10)
dset
```

### Plotting the data

#### 2D spectra

```python
dset.isel(time=slice(None, None, 3)).spec.plot(
    col="time",
    col_wrap=10,
    add_colorbar=False,
    show_theta_labels=False,
    show_radii_labels=False,
)
plt.tight_layout()
```

#### Drifting stats

```python
fig, axs = plt.subplots(1, 2, figsize=(20, 8))

ax = axs[0]
p = ax.scatter(dset.lon, dset.lat, s=100, c=dset.hs, cmap="turbo", alpha=0.5)
plt.colorbar(p, label="Hs (m)", location="top")
ax.set_title("Hs from file")

ax = axs[1]
p = ax.scatter(dset.lon, dset.lat, s=100, c=dset.spec.hs(), cmap="turbo", alpha=0.5)
plt.colorbar(p, label="Hs (m)", location="top")
ax.set_title("Hs from wavespectra")
```

## The spotter reader also supports the Json format

```python
dset = read_spotter("data/spotter_20180214.json")
dset
```

The filetype is inferred from the file name. They can also be explicitly specified if
the file is named differently:

```python
dset = read_spotter("data/spotter_20180214.json", filetype="json", dd=2.0)
dset
```
